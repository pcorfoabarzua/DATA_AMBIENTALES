# Limpieza de datos Macrofauna


```python
import numpy as np 
import pandas as pd
import re 
import datetime
import traceback as tb
import sys



df_macrofauna = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\2_BD_SP_MACROFAUNA_2021.xlsx' , sheet_name= 'MACROFAUNA2021_656-140')

df_macrofauna_quitralco = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\5_BD_SP_MACROFAUNA_Quitralco_2021.xlsx',  sheet_name = 'MACROFAUNA2021_656-140')

#Eliminamos la primera fila de df_macrofauna_quitralco para poder aplicar la funcion 

df_macrofauna_quitralco = df_macrofauna_quitralco.iloc[1: , :]

#Reiniciamos los indices 

df_macrofauna_quitralco = df_macrofauna_quitralco.reset_index(drop=True)


```

```python
df_macrofauna_quitralco
```

```python
def limpieza_macrofauna(df):

    try:

        #Eliminamos las filas innecesarias
        df = df.iloc[13: , :]

        #Creamos una copia del dataset 
        df_macrofauna_copy = df.copy()

        #Reiniciamos los indices del dataset
        df_macrofauna_copy = df_macrofauna_copy.reset_index(drop = True)

        #Dejamos la primera fila como columna 
        df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

        #Eliminamos la primera fila 
        df_macrofauna_copy = df_macrofauna_copy.iloc[1: , :]

        #Reiniciamos los indices 
        df_macrofauna_copy = df_macrofauna_copy.reset_index(drop = True)

        #Creamos una columna de ID 
        df_macrofauna_copy.insert(0 , 'ID' , df_macrofauna_copy.index + 1 )

        #Creamos la lista de columnas 
        columnas = df_macrofauna_copy.columns.to_list()

        #Creamos la lista de columnas fijas 

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'ETAP_PROY', 'JEFE_PROY', 'INST_PROY',
       'NOM_SECTOR', 'COD_REG', 'VARIABLE', 'FECHA_INI', 'FECHA_TER',
       'N_ESTACION', 'LATITUD', 'LONGUITUD', 'PROF_SECT', 'PHYLUM', 'CLASE', 'FAMILIA', 'GEN/ESPECIE']
        
        columnas_variables = [ 'N°_TAXA' , 'NUM_EJEMPL' , 'TIPO_DRAGA' , 'ABUNDANCIA', 'BIOMASA', 'IND_M2', 'GRS_M2']

        #Realizamos un melt para convertir las columnas de variables a filas 

        df_macrofauna_copy = pd.melt(df_macrofauna_copy , id_vars= columnas_fijas , value_vars=columnas_variables , var_name= 'VARIABLE_MUESTREO' , value_name='VALOR')

        #Creamos una funcion para convertir la hora en datetime 

        #Reemplazamos los valores de variables

        df_macrofauna_copy['VARIABLE_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLE_MUESTREO'] == 'ABUNDANCIA' , 'ABUNDANCIA TOTAL' , df_macrofauna_copy['VARIABLE_MUESTREO'])
        df_macrofauna_copy['VARIABLE_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLE_MUESTREO'] == 'N°_TAXA' , 'NUMERO DE TAXA' , df_macrofauna_copy['VARIABLE_MUESTREO'])
        df_macrofauna_copy['VARIABLE_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLE_MUESTREO'] == 'BIOMASA' , 'BIOMASA TOTAL' , df_macrofauna_copy['VARIABLE_MUESTREO'])
        df_macrofauna_copy['VARIABLE_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLE_MUESTREO'] == 'NUM_EJEMPL' , 'NUMERO DE EJEMPLARES' , df_macrofauna_copy['VARIABLE_MUESTREO'])
        

        

        #Cambiamos los formatos de las fechas 

        df_macrofauna_copy['FECHA_INI'] = np.where(df_macrofauna_copy['FECHA_INI'] == '31-11-2021', datetime.datetime(2021 , 11 , 30 , 0 , 0) , df_macrofauna_copy['FECHA_INI'] )
        df_macrofauna_copy['FECHA_TER'] = np.where(df_macrofauna_copy['FECHA_TER'] == '31-11-2021', datetime.datetime(2021 , 11 , 30 , 0 , 0) , df_macrofauna_copy['FECHA_TER'] )
      

        #Forzamos el tipo de FECHA_INI y FECHA_TER a date

        df_macrofauna_copy['FECHA_INI'] =  pd.to_datetime(df_macrofauna_copy['FECHA_INI'])
        df_macrofauna_copy['FECHA_TER'] =  pd.to_datetime(df_macrofauna_copy['FECHA_TER'])
        


        #Cambiamos el formato de las fechas 

        df_macrofauna_copy['FECHA_INI'] = df_macrofauna_copy['FECHA_INI'].dt.strftime('%d/%m/%Y %H:%M:%S')
        df_macrofauna_copy['FECHA_TER'] = df_macrofauna_copy['FECHA_TER'].dt.strftime('%d/%m/%Y %H:%M:%S')
        
        

        return df_macrofauna_copy , columnas , columnas_fijas , columnas_variables

    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_macrofauna: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None #O devuelve un valor que indique el fallo
```

```python
#Utilizamos la funcion para poder limpiar los datos de macrofauna

df_macrofauna_copy , columnas , columnas_fijas , columnas_variables = limpieza_macrofauna(df_macrofauna)
```

```python
#Utilizados la funcion para poder limpiar los datos de macrofauna de quitralco

df_macrofauna_quitralco_copy , columnas , columnas_fijas , columnas_variables = limpieza_macrofauna(df_macrofauna_quitralco)
```

# Limpieza de datos de CTD

```python
df_ctd = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\3_BD_SP_CTD_2021.xlsx' , sheet_name = 'BD_CTD2021_656140')

df_ctd_quitralco = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\6_BD_SP_CTD_Quitralco_2021.xlsx' , sheet_name = 'BD_CTD_QUITRALCO_2021_656140' )

df_ctd_quitralco = df_ctd_quitralco.iloc[1: , :]

df_ctd_quitralco = df_ctd_quitralco.reset_index(drop=True)

```

```python
df_ctd_quitralco
```

```python
def limpieza_ctd(df):

    try:
        
        #Eliminamos las filas innecesarias 
        df = df.iloc[19: , :]

        #Realizamos una copia del dataset
        df_ctd_copy = df.copy()

        #Reiniciamos los indices del dataset
        df_ctd_copy = df_ctd_copy.reset_index(drop = True)

        #Dejamos la primera fila como columna 
        df_ctd_copy.columns = df_ctd_copy.iloc[0]

        #Eliminamos la primera fila del dataset
        df_ctd_copy = df_ctd_copy.iloc[1: , :]

        #Reiniciamos los indices
        df_ctd_copy = df_ctd_copy.reset_index(drop = True)

        #Eliminamos la ultima fila del dataset
        df_ctd_copy = df_ctd_copy.iloc[:-1 , :]

        #Reiniciamos los indices
        df_ctd_copy = df_ctd_copy.reset_index(drop = True)
        
        #Creamos una columna de ID 
        df_ctd_copy.insert(0 , 'ID' , df_ctd_copy.index + 1)

        #Creamos la lista de columnas 
        columnas = df_ctd_copy.columns.to_list()

        #Creamos las listas de columnas fijas y columnas variables 

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'FECHA_INI',  'FECHA_TER', 'EQUIPO', 'LATITUD', 'LONGUITUD', 'PROF_EQ', 'PROF_SECT' ,  'COD_REG', 'ESTACION', 'HORA_INICIO']
        
        columnas_variables = ['TEMPERATUR', 'OXIG_ml/L', 'OXIG_mg/L', 'OXIG_%sat', 'OX_umol/kg', 'SALINIDAD', 'FLUORESCEN', 'DENSIDAD', 'CLOROFILA', 'FEOPIGMEN', 'NITRATO', 'FOSFATO', 'SILICATO']

        #Realizamos el melt para convertir las columnas variables en filas

        df_ctd_copy = pd.melt(df_ctd_copy , id_vars= columnas_fijas , value_vars=columnas_variables , var_name= 'VARIABLE' , value_name='VALOR')

        #Convertir la fecha de inicio utilizando la FECHA_INI y HORA_INI

        # Creamos una funcion para convertir la columna de HORA_INICIO a datetime para luego Crear una columna de fecha y hora 
        def convertir_hora(time_value):
            if isinstance(time_value, str): 
                # Separamos el ":" del texto y lo convertimos a entero y obtenemos dos variables
                horas, minutos = map(int, time_value.split(':'))
                #Retornamos los valores de Horas y minutos
                return datetime.time(horas, minutos)
            else:
                # En caso que ya es datetime.time se deja como esta
                return time_value
            

        # Aplicamos la funcion de convertir hora para dejar todas las HORA_INICIO en datetime.time (Formato Hora)
        df_ctd_copy['HORA_INICIO'] = df_ctd_copy['HORA_INICIO'].apply(convertir_hora)
      

        #Creamos una columna llamada FECHA_INI_CONVERTIDA 
        df_ctd_copy['FECHA_INI_CONVERTIDA'] =  pd.to_datetime(df_ctd_copy['FECHA_INI'] , format= '%d-%m-%Y').dt.strftime("%d/%m/%Y")

        #Creamos una columna llamada FECHA_INI_CONVERTIDA 
        df_ctd_copy['FECHA_TER_CONVERTIDA'] =  pd.to_datetime(df_ctd_copy['FECHA_TER'] , format= '%d-%m-%Y').dt.strftime("%d/%m/%Y")


        
        #Luego creamos una nueva columna para combinar la FECHA_INI con la Hora

        df_ctd_copy['FECHA_HORA_COMBINADA_DT'] = df_ctd_copy.apply(
        lambda row: datetime.datetime.strptime(row['FECHA_INI_CONVERTIDA'], "%d/%m/%Y").replace(
            hour=row['HORA_INICIO'].hour,
            minute=row['HORA_INICIO'].minute,
            second=row['HORA_INICIO'].second
        ),
        axis=1
    )
        
        #Reemplazamos los valores de NOMB_PROY 
        df_ctd_copy['NOMB_PROY'] = np.where(df_ctd_copy['NOMB_PROY'] == 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2021 -2022' , 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2021-2022' , df_ctd_copy['NOMB_PROY'])

        #Renombramos las columnas 
        df_ctd_copy = df_ctd_copy.rename(columns={'LONGUITUD' : 'LONGITUD'})

        #Renombramos los nombres de las VARIABLES
        df_ctd_copy['VARIABLE'] = np.where(df_ctd_copy['VARIABLE'] == 'TEMPERATUR' , 'TEMPERATURA' , df_ctd_copy['VARIABLE'])
        #Renombramos los nombres de las VARIABLES
        df_ctd_copy['VARIABLE'] = np.where(df_ctd_copy['VARIABLE'] == 'FLUORESCEN' , 'FLUORESCENCIA' , df_ctd_copy['VARIABLE'])


        #Aplicamos un UPPER a la columna VARIABLE
        df_ctd_copy['VARIABLE'] = df_ctd_copy['VARIABLE'].str.upper()


        return df_ctd_copy , columnas , columnas_fijas , columnas_variables


    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_ctd: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None #O devuelve un valor que indique el fallo



```

```python
#Utilizamos la funcion para limpiar los datos de CTD

df_ctd_copy , columnas , columnas_fijas , columnas_variables= limpieza_ctd(df_ctd)
```

# Limpieza de datos MUESTREO REGULAR y MUESTREO REGULAR Quiltralco

```python
df_muestreo_regular = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\1_BD_SP_MuestreoRegular_2021.xlsx' , sheet_name = 'BD_MREGULAR_IFOP2021')

df_muestreo_regular_quiltralco = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2021_P656140\4_BD_SP_SEDIMENTO_Quitralco_2021.xlsx' , sheet_name = 'BD_SedimentosQUITRALCO_IFOP2021')
```

```python
def limpieza_muestreo_regular(df):

    try:

        #Eliminamos filas innecesarias 
        df = df.iloc[17: , :]

        #Reiniciamos los indices del dataset
        df = df.reset_index(drop = True)

        #Dejamos la primera fila como columnas del dataset
        df.columns = df.iloc[0]

        #Eliminamos la primera fila 
        df =  df.iloc[1: , :]

        #Reiniciamos los indices del dataset 
        df = df.reset_index(drop = True)

        #Creamos una copia del dataset
        df_muestreo_regular_copy = df.copy()

        #Creamos la columna ID 
        df_muestreo_regular_copy.insert(0 , 'ID' , df_muestreo_regular_copy.index + 1 )

        #Renombramos las columnas de NOM_SECTOR 

        df_muestreo_regular_copy.columns.values[6] = 'NOM_SECTOR_1'
        df_muestreo_regular_copy.columns.values[18] = 'NOM_SECTOR_2'

        #Eliminamos la columna de NOM_SECTOR_2 que está de más

        df_muestreo_regular_copy.drop(columns=['NOM_SECTOR_2'], inplace=True)

        #Renombramos la columna NOM_SECTOR_1 a NOM_SECTOR

        df_muestreo_regular_copy.columns.values[6] = 'NOM_SECTOR'

        #Creamos la lista de columnas 
        columnas = df_muestreo_regular_copy.columns.to_list()

        #Creamos las columnas fijas y columnas Variables

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'ETAP_PROY', 'JEFE_PROY', 'INST_PROY', 'NOM_SECTOR', 'COD_REG', 'VARIABLE', 'FECHA_INI', 'FECHA_TER','HORA_INI', 'HORA_FIN', 'N_ESTACION', 'LATITUD', 'LONGUITUD','PROF_SECT' ,'OBSERV']

        columnas_variables = [ 'PROF_VAR', 'Ph_REPL_1', 'Ph_REPL_2', 'Ph_REPL_3', 'TEMP_REPL_1', 'TEMP_REPL_2', 'TEMP_REPL_3', 'REDOX_REPL_1','REDOX_REPL_2', 'REDOX_REPL_3', 'SULF_REPL_1', 'SULF_REPL_2','SULF_REPL_3', 'MAT_ORG', 'ABUNDANCIA', 'N°_TAXA', 'BIOM_TOTAL','DIV_SHANNON', 'UNIF_PIELOU', 'AMBI']

        #Realizamos un melt para convertir las columnas de variables en filas

        df_muestreo_regular_copy = pd.melt(df_muestreo_regular_copy , id_vars= columnas_fijas , value_vars= columnas_variables , var_name= 'VARIABLE_MUESTREO' , value_name='VALOR' )

    	#Se van a eliminar todas las filas que la FECHA_INI es "sd" ya que no sirven para el proposito de la BD. Se eliminaron 20 registros
        df_muestreo_regular_limpio = df_muestreo_regular_copy[df_muestreo_regular_copy['FECHA_INI'] != 'sd']


        #Se van a eliminar todos los registros que tienen la FECHA_INI DE 31/11 porque esa fecha no existe 
        df_muestreo_regular_limpio = df_muestreo_regular_limpio[df_muestreo_regular_limpio['FECHA_INI'] != '31-11-2021']

        #Cambiamos los tipos de FECHA_INI y FECHA_FIN a datetime

        df_muestreo_regular_limpio['FECHA_INI'] = pd.to_datetime(df_muestreo_regular_limpio['FECHA_INI'])
        df_muestreo_regular_limpio['FECHA_TER'] = pd.to_datetime(df_muestreo_regular_limpio['FECHA_TER'])

        #FECHA INI CONVERTIDA A STRING 
        df_muestreo_regular_limpio['FECHA_INI_STR'] =  df_muestreo_regular_limpio['FECHA_INI'].dt.strftime("%d/%m/%Y")
        df_muestreo_regular_limpio['FECHA_FIN_STR'] = df_muestreo_regular_limpio['FECHA_TER'].dt.strftime("%d/%m/%Y")
        
        #Para poder combinar las fechas y horas hay que convertir los datetime a string mediante df['FECHA'].dt.strftime("%d/%m/%Y")
        

        # Realizamos una funcion lambda para poder combinar la FECHA_INI_STR con la hora y convertirlo a datetime

        df_muestreo_regular_limpio['FECHA_INICIO'] = df_muestreo_regular_limpio.apply(
        lambda row: datetime.datetime.strptime(row['FECHA_INI_STR'], "%d/%m/%Y").replace(
            hour=row['HORA_INI'].hour,
            minute=row['HORA_INI'].minute,
            second=row['HORA_INI'].second
            ),
            axis=1
        )

        #Realizamos una funcion lambda para conbinar la FECHA_FIN_STR con la HORA_FIN

        df_muestreo_regular_limpio['FECHA_TERMINO'] = df_muestreo_regular_limpio.apply(
        lambda row: datetime.datetime.strptime(row['FECHA_FIN_STR'], "%d/%m/%Y").replace(
            hour=row['HORA_FIN'].hour,
            minute=row['HORA_FIN'].minute,
            second=row['HORA_FIN'].second
            ),
            axis=1
        )

        #Aplicamos un UPPER a la columna de VARIABLE_MUESTREO

        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = df_muestreo_regular_limpio['VARIABLE_MUESTREO'].str.upper()

        #Cambiamos los nombres de las VARIABLE_MUESTREO

        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'MAT_ORG' , 'MATERIA ORGANICA' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])
        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'ABUNDANCIA' , 'ABUNDANCIA TOTAL' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])
        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'N°_TAXA' , 'NUMERO DE TAXA' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])
        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'BIOM_TOTAL' , 'BIOMASA TOTAL' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])
        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'DIV_SHANNON' , 'DISIDAD DE SHANNON' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])
        df_muestreo_regular_limpio['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_limpio['VARIABLE_MUESTREO'] == 'UNIF_PIELOU' , 'UNIFORMIDAD DE PIELOU' , df_muestreo_regular_limpio['VARIABLE_MUESTREO'])


        #Modificamos el formato de FECHA_INICIO y FECHA_TERMINO

        df_muestreo_regular_limpio['FECHA_INICIO'] = df_muestreo_regular_limpio['FECHA_INICIO'].dt.strftime('%d/%m/%Y %H:%M:%S')
        df_muestreo_regular_limpio['FECHA_TERMINO'] = df_muestreo_regular_limpio['FECHA_TERMINO'].dt.strftime('%d/%m/%Y %H:%M:%S')


        #Modificamos la columna VALOR de sd , '  ****' a np.nan

        df_muestreo_regular_limpio['VALOR'] = np.where(df_muestreo_regular_limpio['VALOR'] == 'sd' , np.nan , df_muestreo_regular_limpio['VALOR'])
        df_muestreo_regular_limpio['VALOR'] = np.where(df_muestreo_regular_limpio['VALOR'] == '  ****' , np.nan , df_muestreo_regular_limpio['VALOR'])

        
        return df_muestreo_regular_copy  ,  df_muestreo_regular_limpio  ,columnas , columnas_fijas , columnas_variables
    
    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_muestreo_regular: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None #O devuelve un valor que indique el fallo
```

```python
#Aplicamos la funcion para limpiar los datos 

df_muestreo_regular_copy , df_muestreo_regular_limpio , columnas , columnas_fijas , columnas_variables= limpieza_muestreo_regular(df_muestreo_regular)
```
