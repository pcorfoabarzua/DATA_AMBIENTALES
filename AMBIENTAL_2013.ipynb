# Limpieza de datos Proyecto Ambientales año 2013 

    Limpieza de MACROFAUNA



```python
import re 
import numpy as np 
import pandas as pd 
import datetime

# Leemos los datos de Macrofauna
df_macrofauna = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2013_P656\2_BD_MACROFAUNA_2013.xlsx' , sheet_name= '2013')

```

```python
def limpieza_macrofauna(df):

    #Eliminamos los datos innecesarios 

    df = df.iloc[15: , :21]

    #Creamos una copia del dataframe 

    df_macrofauna_copy = df.copy()

     # Reiniciamos los index
    df_macrofauna_copy =  df_macrofauna_copy.reset_index()

    #Eliminamos la columna index 
    df_macrofauna_copy =  df_macrofauna_copy.iloc[: , 1:]

    #Dejamos la primera fila como columnas 
    df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

     #Eliminamos la primera fila 
    df_macrofauna_copy =  df_macrofauna_copy.iloc[1: , :]

    # Creamos una lista de las columnas
    columnas = df_macrofauna_copy.columns.tolist()

    # Realizamos cambios en la lista de columnas 

    columnas[17] = 'AMBI INV'
    columnas[18] = 'CLASIFICACIÓN DE PERTURBACIÓN INV'
    columnas[19] = 'AMBI VER'
    columnas[20] =  'CLASIFICACIÓN DE PERTURBACIÓN VER'

     # Realizamos dos ciclos para agregar a los campos si son pertenecientes a temporada invierno o verano

    for indice , columna in enumerate(columnas):

        if indice >= 7 and indice < 17:

            print(f"La columna '{columna}' esta en la posicion {indice}")

            if indice % 2 == 0:

                print(f"La columna '{columna} se debe agregar '{columnas[indice - 1]}' y VER para convertirse en  '{columnas[indice - 1]} VER' ")

                columnas[indice] = columnas[indice - 1] + ' VER' 


    for indice , columna in enumerate(columnas):

        if indice >= 7 and indice < 17:

            print(f"La columna '{columna}' esta en la posicion {indice}")

            if indice % 2 != 0:

                print(f"La columna '{columna} se debe agregar INV para convertirse en  '{columnas[indice]} INV ' ")

                columnas[indice] = columnas[indice] + ' INV' 


    # Utilizamos expresiones regulares para eliminar los espacios sobrantes de un texto de una lista

    columnas_limpias = [re.sub(r'\s+' , ' ' ,  columna) for columna in columnas]


    # Ahora reemplazamos las columnas del dataframe por la lista de columnas 

    df_macrofauna_copy.columns = columnas_limpias

    # Ahora eliminamos la primera fila del dataframe 

    df_macrofauna_copy = df_macrofauna_copy.iloc[1: , :]

      # Reiniciamos los index
    df_macrofauna_copy =  df_macrofauna_copy.reset_index()

    #Eliminamos la columna index 
    df_macrofauna_copy =  df_macrofauna_copy.iloc[: , 1:]
    

    #Creamos una columna de ID en la posicion 0 
    df_macrofauna_copy.insert(0 , 'ID' , df_macrofauna_copy.index + 1 )


    #Volvemos a crear las columnas limpias
    columnas_limpias = df_macrofauna_copy.columns.tolist()


    # Creamos una lista con las columnas fijas 

    columnas_fijas = columnas_limpias[:8]

    columnas_variables = columnas_limpias[9:]



    # Ahora realizamos un melt para convertir las columnas de variables a filas 

    df_macrofauna_copy = pd.melt(df_macrofauna_copy , columnas_fijas , columnas_variables , var_name= 'VARIABLE' , value_name= 'VALOR')


    #Creamos una columna adicional para diferenciar que es INVERNO O VERANO 

    # Creamos una funcion para extraer los INV o VER de las Variables 

    def obtener_ultimo_tres(texto):
        if isinstance(texto, str) and len(texto) >= 3:
            return texto[-3:]
        else:
            return None  # O algún otro valor por defecto si la cadena es muy corta o no es string

    # Aplica la función a la columna y crea una nueva columna
    df_macrofauna_copy['ESTACION_DEL_ANIO'] = df_macrofauna_copy['VARIABLE'].apply(obtener_ultimo_tres)


    # Creamos una funcion para eliminar los INV y VER de la columna VAriable 

    def eliminar_inv_ver(texto):

        if isinstance(texto, str):
         texto = texto.replace('INV', '')
         texto = texto.replace('VER', '')
        return texto


    #Aplicamos la funcion para eliminar los INV y VER de la columna VARIABLE
    df_macrofauna_copy['VARIABLE'] = df_macrofauna_copy['VARIABLE'].apply(eliminar_inv_ver)

    #Eliminamos los espacios en blanco al final del texto 
    df_macrofauna_copy['ZONA'] = df_macrofauna_copy['ZONA'].str.rstrip()
    df_macrofauna_copy['VARIABLE'] = df_macrofauna_copy['VARIABLE'].str.rstrip()

    #Reemplazamos los valores de 'sp' de la columna VALOR a null
    df_macrofauna_copy['VALOR'] = np.where(df_macrofauna_copy['VALOR'] == 'sd' , np.nan ,  df_macrofauna_copy['VALOR'])

    # Creamos un diccionario para poder mapear las zonas con las regiones de Chile 

    mapeo_regiones = {'Estuario' : 'X' , 'Reloncavi' : 'X' , 'Calbuco' : 'X' , 'Maullin' : 'X' , 'Ancud' : 'X' , 'Chacao' : 'X' , 'Butachauques' : 'X' , 'Chaulinec' : 'X' , 'Tenaun' : 'X' , 'Chiloe central'  : 'X' ,
                  'Desertores' : 'X' , 'Quellon' : 'X' ,  'Hornopiren' : 'X' , 'Ayacara' : 'X' , 'Chaiten' : 'X' ,  'Guaitecas' : 'XI' , 'Moraleda' : 'XI' , 'Puyuhuapi' : 'XI' , 'Aysen' : 'XI' , 'Coquimbo' : 'IV' , 'Atacama' : 'III' , 'Pto. Natales' : 'XII'   }

    #Creamos la columna COD_REGION utilizando la Zona con el mapeo_regiones

    df_macrofauna_copy['COD_REGION'] = df_macrofauna_copy['ZONA'].map(mapeo_regiones)

    #Insertamos una columna de año 
    df_macrofauna_copy.insert(1 , 'AÑO' , 2013)


    #Reemplazamos los valores de variables para eliminar la unidad de la variable 

    df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'ABUNDANCIA TOTAL (ind/m2)' , 'ABUNDANCIA TOTAL' , df_macrofauna_copy['VARIABLE'])
    df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'BIOMASA TOTAL(g/m2)' , 'BIOMASA TOTAL' , df_macrofauna_copy['VARIABLE'])

    #Reemplazamos los valores de NOMB_PROY 

    df_macrofauna_copy['NOMB_PROY'] = np.where(df_macrofauna_copy['NOMB_PROY'] == 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2013 -2014' , 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2013-2014' , df_macrofauna_copy['NOMB_PROY'])
    df_macrofauna_copy['NOMB_PROY'] = np.where(df_macrofauna_copy['NOMB_PROY'] == 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2012 -2013' , 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2012-2013' , df_macrofauna_copy['NOMB_PROY'])


    return df_macrofauna_copy ,  columnas_limpias


```

```python
df_macrofauna_copy  , columnas = limpieza_macrofauna(df_macrofauna)
```

# Limpieza de CTD 2013

```python
df_ctd = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2013_P656\3_BD_CTD_2013.xlsx' , sheet_name='2013')
```

```python
df_ctd
```

```python
def limpieza_ctd(df):

    # Eliminamos las filas innecesarias 
    df = df.iloc[10: , :]

    #Creamos una copia del dataframe 
    df_ctd_copy = df.copy()

    # Reiniciamos los index
    df_ctd_copy =  df_ctd_copy.reset_index()

    #Eliminamos la columna index 
    df_ctd_copy =  df_ctd_copy.iloc[: , 1:]
    
    #Dejamos la primera fila como columna
    df_ctd_copy.columns =  df_ctd_copy.iloc[0]

    #Eliminamos las dos primera fila 
    df_ctd_copy = df_ctd_copy.iloc[2: , :]

    # Reiniciamos los index
    df_ctd_copy =  df_ctd_copy.reset_index()

    #Eliminamos la columna index 
    df_ctd_copy =  df_ctd_copy.iloc[: , 1:]

    # Creamos una columna de FECHA en la posicion 9 del dataframe
    df_ctd_copy.insert(10 , 'FECHA' , df_ctd_copy.apply(lambda row: datetime.datetime(row['YEAR'], row['MONTH'], row['DAY']), axis=1) )  

    #Creamos una columna de ID en la posicion 0 
    df_ctd_copy.insert(0 , 'ID' , df_ctd_copy.index + 1 )

    #Formateamos la FECHA  a DD/MM/YYYY 
    # Formatear la fecha y hora al formato deseado
    df_ctd_copy['FECHA'] = df_ctd_copy['FECHA'].dt.strftime('%d/%m/%Y %H:%M:%S')

    #Renombramos las columnas de variables
    df_ctd_copy = df_ctd_copy.rename(columns= {  'STATION' : 'ESTACION'  , 'CRUISE'  : 'SEGUIMIENTO' , 'LATITUDE' : 'LATITUD' , 'LONGITUDE' : 'LONGITUD'  ,  'TEMPERATURE_(°C)' : 'TEMPERATURA °C' ,  'OXIGENO_DIS_(ML/L)' : 'OXIGENO_DIS_ML/L ML/L' , 'OXIGENO_DIS_(MG/L)' : 'OXIGENO_DIS_MG/L MG/L' , 'OXIGENO_DIS_(%SAT)' : 'OXIGENO_DIS_SAT %' , 'OXIGENO_DIS_(UMOL/KG)' : 'OXIGENO_DIS_UMOLKG UMOL/KG' , 'SALINITY_(PSU)' : 'SALINIDAD PSU' , 'DENSIDAD_(KG/M3)' : 'DENSIDAD KG/M3' , 'FLUORESCENCIA_(MG/M-3)' : 'FLUORESCENCIA MG/M-3' , 'DEPHT_(M)' : 'PROFUNDIDAD'})

    #Creamos una lista con las columnas del dataframe
    columnas = df_ctd_copy.columns.tolist()

    
    #Obtenemos las columnas fijas y las columnas Variables 
    columnas_fijas = columnas[:12]
    columnas_variables = columnas[12:]
    

    #Realizamos un melt para convertir las columnas de Variables a Filas 

    # Dataframe 
    # id_vars : columnas_fijas
    # value_vars : columnas_variables
    # var_name : Nombre de la nueva columna 
    # value_name : Nombre de la nueva columna de valores 

    df_ctd_copy  = pd.melt(df_ctd_copy , id_vars= columnas_fijas , var_name= 'VARIABLE' , value_vars= columnas_variables , value_name= 'VALOR' )

    #Volvemos a crear una lista con las columnas del dataframe
    columnas = df_ctd_copy.columns.tolist()

    #Creamos una columna llamada Unidad obteniendo lo que viene despues del espacio en la columna VARIABLE
    df_ctd_copy.insert(12 , 'UNIDAD' , df_ctd_copy['VARIABLE'].str.split(' ', n=1).str[1])

    #Luego eliminamos las Unidades de la columna Variable 
    df_ctd_copy['VARIABLE'] =  df_ctd_copy['VARIABLE'].str.split(' ' , n=1).str[0]

    #Dejamos todas las estaciones en MAYUSCULA
    df_ctd_copy['ESTACION'] =  df_ctd_copy['ESTACION'].str.upper()

    #Modificamos la columna de NOMB_PROY
    df_ctd_copy['NOMB_PROY'] = np.where(df_ctd_copy['NOMB_PROY'] == 'ESTUDIO DEL DESEMPEÑO AMBIENTAL DE LA ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO, 2013-2014' , 'ESTUDIO DEL DESEMPEÑO AMBIENTAL DE LA ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2013-2014' ,  df_ctd_copy['NOMB_PROY'])

    
    # Creamos un diccionario de Mapeo para los sectores 

    # Crear un diccionario de mapeo para tus asignaciones
    mapeo_sectores = {
            
            "TEN" :  "Tenaun" ,
            "CHAU" : "Chaulinec" ,
            "CHIC" : "Chiloe central" ,
            "CHAC" : "Chacao" ,
            "QUELL" : "Quellon",
            "BUT" :  "Butachauques",
            "ANC" : "Ancud"

    }


        
    mapeo_regiones = {'Estuario' : 'X' , 'Reloncavi' : 'X' , 'Calbuco' : 'X' , 'Maullin' : 'X' , 'Ancud' : 'X' , 'Chacao' : 'X' , 'Butachauques' : 'X' , 'Chaulinec' : 'X' , 'Tenaun' : 'X' , 'Chiloe central'  : 'X' ,
                  'Desertores' : 'X' , 'Quellon' : 'X' ,  'Hornopiren' : 'X' , 'Ayacara' : 'X' , 'Chaiten' : 'X' ,  'Guaitecas' : 'XI' , 'Moraleda' : 'XI' , 'Puyuhuapi' : 'XI' , 'Aysen' : 'XI' , 'Coquimbo' : 'IV' , 'Atacama' : 'III' , 'Pto. Natales' : 'XII' , 'Bio-Bio' : 'VIII' }


    # Extraemos los Números de la columna ESTACION para crear una la columna N_ESTACION
    df_ctd_copy['N_ESTACION'] = df_ctd_copy['ESTACION'].str.extract(r'(\d+)')

    #Obtenemos el código utilizando la expresion regular 
    df_ctd_copy['CODIGO_EXTRAIDO'] = df_ctd_copy['ESTACION'].str.extract(r'([A-Z]{3,5})')

    #Luego utilizando el diccionario de mapeo creamos la columna de SECTOR utilizando la funcion map

    df_ctd_copy['SECTOR'] = df_ctd_copy['CODIGO_EXTRAIDO'].map(mapeo_sectores)

    #Utilizamos el diccionario de mapeo de regiones para crear la columna COD_REGION

    df_ctd_copy['COD_REGION'] = df_ctd_copy['SECTOR'].map(mapeo_regiones)


    return df_ctd_copy , columnas , columnas_fijas , columnas_variables

```

```python
df_ctd_copy , columnas , columnas_fijas , columnas_variables= limpieza_ctd(df_ctd)
```
