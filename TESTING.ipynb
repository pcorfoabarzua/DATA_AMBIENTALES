```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'texto_con_tildes': ['camión', 'canción', 'árbol', 'teléfono', 'mamá', 'pingüino']}
df = pd.DataFrame(data)

# Mostrar el DataFrame original
print("DataFrame Original:")
print(df)
print("-" * 30)

# Aplicar la limpieza
df['texto_sin_tildes'] = df['texto_con_tildes'].str.normalize('NFD').str.replace(r'[\u0300-\u036f]', '', regex=True)

# Mostrar el DataFrame modificado
print("DataFrame Modificado:")
print(df)
```
