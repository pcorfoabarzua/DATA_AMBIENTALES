```python
def limpieza_macrofauna(df):

    try:

        #Eliminamos las fil as innecesarias
        df =  df.iloc[13: , :]

        #Reiniciamos los indices del dataframe y eliminamos la columna de index 
        df =  df.reset_index(drop = True)

        #Creamos una copia del dataframe
        df_macrofauna_copy = df.copy()

        #Dejamos la primera fila como columna
        df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

        #Eliminamos la primera fila del dataset 
        df_macrofauna_copy = df_macrofauna_copy.iloc[1: , :]

        #Reiniciamos los indices del dataframe
        df_macrofauna_copy = df_macrofauna_copy.reset_index(drop =  True)

        #Agregamos la columna ID 
        df_macrofauna_copy.insert(0 , 'ID' , df_macrofauna_copy.index + 1)

        #Creamos la lista de columnas 
        columnas = df_macrofauna_copy.columns.to_list()

        #Creamos las columnas fijas y las columnas variables

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'ETAP_PROY', 'JEFE_PROY', 'INST_PROY', 'NOM_SECTOR', 'COD_REG', 'FECHA_INI', 'FECHA_TER', 'N_ESTACION', 'LATITUD', 'LONGUITUD', 'PROF_SECT','PHYLUM', 'CLASE/ORD', 'FAMILIA', 'GEN/ESPECIE']


        columnas_variables = ['NUM_EJEMPL', 'TIPO_DRAGA', 'ABUNDANCIA', 'BIOMASA', 'IND_M2', 'GRS_M2']

        #Realizamos un melt para convertir las columnas de variables a filas 

        df_macrofauna_copy = pd.melt(df_macrofauna_copy , id_vars= columnas_fijas , value_vars=columnas_variables , var_name= 'VARIABLES_MUESTREO' , value_name= 'VALOR')







        #Renombramos los nombres de VARIABLES_MUESTREO

        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'ABUNDANCIA' , 'ABUNDANCIA TOTAL' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'BIOMASA' , 'BIOMASA TOTAL' , df_macrofauna_copy['VARIABLES_MUESTREO'])


        #Reemplazamos todos los valores de NOMB_PROY

        df_macrofauna_copy['NOMB_PROY'] = 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2019-2020'

        #Renombramos la columnas 

        df_macrofauna_copy = df_macrofauna_copy.rename(columns = {"LONGUITUD" : "LONGITUD" , "GEN/ESPECIE" : "TAXA" , "CLASE/ORD" : "CLASE"})

        # Modificamos los formatos de las fechas a DD/MM/YYYY 24HH:MI:SS

        df_macrofauna_copy['FECHA_INI'] = df_macrofauna_copy['FECHA_INI'].dt.strftime('%d/%m/%Y %H:%M:%S')

        df_macrofauna_copy['FECHA_TER'] = df_macrofauna_copy['FECHA_TER'].dt.strftime('%d/%m/%Y %H:%M:%S')

        
         #Modificamos las columnas de Variable 

        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'N°_TAXA', 'NUMERO DE TAXA' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'NUM_EJEMPL', 'NUMERO DE EJEMPLARES' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'TIPO_DRAGA', 'TIPO DRAGA' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'TIPO_DRAGA', 'TIPO DRAGA' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'ABUNDANCIA' , 'ABUNDANCIA TOTAL' , df_macrofauna_copy['VARIABLES_MUESTREO'])
        df_macrofauna_copy['VARIABLES_MUESTREO'] = np.where(df_macrofauna_copy['VARIABLES_MUESTREO'] == 'BIOMASA' , 'BIOMASA TOTAL' , df_macrofauna_copy['VARIABLES_MUESTREO'])



        return df_macrofauna_copy , columnas

    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_ctd: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
        return None #O devuelve un valor que indique el fallo

```

```python
def limpieza_ctd(df):
    
    try:
        # Eliminamos las filas innecesarias
        df = df.iloc[18:, :]

        # Reiniciamos los indices
        df = df.reset_index(drop=True)

        # Creamos una copia del dataset
        df_ctd_copy = df.copy()

        # Dejamos la primera fila como columna
        df_ctd_copy.columns = df_ctd_copy.iloc[0]

        # Eliminamos la primera fila
        df_ctd_copy = df_ctd_copy.iloc[1:, :]

        # Reiniciamos los indices
        df_ctd_copy = df_ctd_copy.reset_index(drop=True)

        # Agregamos una columna de ID para identificar los datos antes de realizar el Melt
        df_ctd_copy.insert(0, 'ID', df_ctd_copy.index + 1)

        # Renombramos las columnas
        df_ctd_copy = df_ctd_copy.rename(columns={  'ESTACION' : 'N_ESTACION' ,  'LONGUITUD': 'LONGITUD', 'TEMPERATUR': 'TEMPERATURA', 'HORA_DATO  ': 'HORA_DATO', 'OXIG_ml/L': 'OXIGENO_DIS_ML/L', 'OXIG_mg/L': 'OXIGENO_DIS_MG/L', 'OXIG_%sat': 'OXIGENO_DIS_SAT', 'OX_umol/kg': 'OXIGENO_DIS_UMOLKG', 'FLUORESCEN': 'FLUORESCENCIA'})

        # Creamos las columnas fijas y columnas de variables
        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'FECHA_INI', 'FECHA_TER', 'EQUIPO', 'LATITUD', 'LONGITUD', 'PROF_EQ', 'PROF_SECT', 'COD_REG', 'OBSERV', 'SHAPE', 'N_ESTACION', 'HORA_DATO']
        columnas_variables = ['TEMPERATURA', 'OXIGENO_DIS_ML/L', 'OXIGENO_DIS_MG/L', 'OXIGENO_DIS_SAT', 'OXIGENO_DIS_UMOLKG', 'FLUORESCENCIA', 'SALINIDAD', 'DENSIDAD', 'CLOROFILA', 'FEOPIGMEN', 'NITRATO', 'FOSFATO', 'SILICATO']

        """
        Realizamos un melt para convertir las columnas de Variables a filas

            Necesita:
            1. Dataframe a cambiar
            2. id_vars : Lista o Tupla de las columnas Fijas
            3. value_vars : Lista o Tupla de las columnas de Variables
            4. var_name : Nombre de la nueva columna de las Variables
            5. value_name : Nombre de la columna con los valores de la columna de Variables

        """

        df_ctd_copy = pd.melt(df_ctd_copy, id_vars=columnas_fijas, value_vars=columnas_variables, var_name='VARIABLE', value_name='VALOR')

        # Creamos una columna de fecha

        # df_ctd_copy.insert(10 , 'FECHA_DATO' , df_ctd_copy.apply(lambda row: datetime.datetime(row['FECHA_TER']), axis=1) )

        # Convertimos la HORA_DATO a string

        df_ctd_copy['HORA_DATO'] = df_ctd_copy['HORA_DATO'].astype(str)

        # Remplazamos los 'nan'

        df_ctd_copy['HORA_DATO'] = np.where(df_ctd_copy['HORA_DATO'] == 'nan', '00:00:00', df_ctd_copy['HORA_DATO'])

        # Creamos un campo de FECHA_DATO
        df_ctd_copy['FECHA_DATO'] = df_ctd_copy['FECHA_TER'].dt.strftime('%d/%m/%Y') + ' ' + df_ctd_copy['HORA_DATO']

        # Realizamos un formateo a los campos de fechas
        df_ctd_copy['FECHA_INI'] = pd.to_datetime(df_ctd_copy['FECHA_INI'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
        df_ctd_copy['FECHA_TER'] = pd.to_datetime(df_ctd_copy['FECHA_TER'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
        df_ctd_copy['FECHA_DATO'] = pd.to_datetime(df_ctd_copy['FECHA_DATO'], format='%d/%m/%Y %H:%M:%S', errors='coerce')

        #



    
        return df_ctd_copy, columnas_fijas, columnas_variables

    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_ctd: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
        return None, None, None  # O devuelve un valor que indique el fallo

```

```python
def limpieza_fisicos_quimicos(df):

    try:
        
        #Eliminamos las filas innecesarias
        df = df.iloc[12: , :]

        #Creamos una copia del dataset
        df_fq_copy = df.copy()

        #Reiniciamos los index
        df_fq_copy = df_fq_copy.reset_index(drop = True)

        #Dejamos la primera fila como columna 
        df_fq_copy.columns = df_fq_copy.iloc[0]

        #Eliminamos la primera fila 
        df_fq_copy = df_fq_copy.iloc[1: , :]

        #Reiniciamos los index
        df_fq_copy = df_fq_copy.reset_index(drop = True)


        #Creamos una columna de ID 
        df_fq_copy.insert(0 , 'ID' , df_fq_copy.index + 1 )

        #Remplazamos los nombres de las columnas
        df_fq_copy = df_fq_copy.rename(columns = {'CLOROF_ a ' : 'CLOROF_A' , 'LONGUITUD' : 'LONGITUD'})


        #Creamos las listas de columnas 
        columnas= df_fq_copy.columns.to_list()

        #Creamos las columnas fijas para transformar los equipos
        columnas_equipos = [columnas[8] , columnas[13] , columnas[16] , columnas[19] , columnas[22]]

        #Creamos las columnas de Prof_Equipos

        columnas_prof = [columnas[9] , columnas[14] , columnas[17] , columnas[20] , columnas[23]]

        #Creamos las columnas fijas para convertir las columnas de equipos

        columnas_fijas = []

        for columna in columnas:

            if columna not in columnas_equipos:

                columnas_fijas.append(columna)


        #Realizamos un melt para convertir las columnas de equipos a filas 

        df_fq_copy = pd.melt(df_fq_copy , id_vars = columnas_fijas , value_vars= columnas_equipos , var_name= 'COL_EQUIPO' , value_name= 'EQUIPO')

        #Eliminamos la columna de COL_EQUIPO

        df_fq_copy = df_fq_copy[['ID', 'NOMB_PROY', 'COD_PROY', 'FECHA_INI', 'FECHA_TER', 'LATITUD', 'LONGITUD', 'ESTACION', 'PROF_EQU_1', 'PROF_SEC', 'TEMPERATURA', 'Ph', 'PROF_EQU_2', 'SALINIDAD', 'PROF_EQU_3', 'OXIGE_DIS ', 'PROF_EQU_4',
                                'TRANSPAREN', 'PROF_EQU_5', 'LUZ_PAR', 'CLOROF_A', 'CLOROF_ACT',  'FEOPIGM', 'FEOPIGM_%', 'MAT_INORGA', 'MAT_ORGANI', 'NITRATO','FOSFATO', 'SILICATO', 'COD_REG', 'OBSERV', 'SHAPE', 'EQUIPO']]


        #Creamos las listas de columnas 
        columnas = df_fq_copy.columns.to_list()

    
        #Aplicamos lo mismo para las Variables 

        columnas_variables = [columnas[8] ,  columnas[10] , columnas[11] , columnas[12] , columnas[13] , columnas[14] , columnas[15] , columnas[16] , columnas[17] , columnas[18] , 
                              columnas[19] , columnas[20] , columnas[21] , columnas[22] , columnas[23] , columnas[24] , columnas[25] , columnas[26] , columnas[27] , columnas[28]   ]
        

         #Creamos las columnas fijas para convertir las columnas de variables

        columnas_fijas = []

        for columna in columnas:

            if columna not in columnas_variables:

                columnas_fijas.append(columna)


        #Realizamos un melt para convertir las columnas de variables 

        df_fq_copy = pd.melt(df_fq_copy , id_vars= columnas_fijas ,  value_vars= columnas_variables , var_name= 'VARIABLE' , value_name='VALOR')
        

        return df_fq_copy , columnas  ,  columnas_equipos , columnas_fijas , columnas_prof , columnas_variables


    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_muestreo_regular: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))

```
