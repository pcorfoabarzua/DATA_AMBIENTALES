# Limpieza de datos de Ambiental 2020

    MACROFAUNA

```python
import numpy as np 
import pandas as pd
import re 
import datetime
import traceback as tb
import sys

df_macrofauna = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2020_P656130\2_BD_MACROFAUNA_2020.xlsx' , sheet_name= 'BD_MACRO_2020')
```

```python
df_macrofauna
```

```python
def limpieza_macrofauna(df):

    try:
        
        #eliminamos las filas innecesarias
        df = df.iloc[13: , :]

        #Creamos una copia del dataset
        df_macrofauna_copy = df.copy()

        #Reiniciamos los indices del dataset
        df_macrofauna_copy = df_macrofauna_copy.reset_index(drop= True)

        #Dejamos la primera fila como columna 
        df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

        #Eliminamos la primera fila del dataset
        df_macrofauna_copy =  df_macrofauna_copy.iloc[1: , :]

        #Reiniciamos los indices del dataset
        df_macrofauna_copy = df_macrofauna_copy.reset_index(drop= True)

        #Creamos la columna de ID 
        df_macrofauna_copy.insert(0 , 'ID' , df_macrofauna_copy.index + 1 )

        #Obtenermos las columnas del dataset

        columnas = df_macrofauna_copy.columns.to_list()

        #Columnas de Variables 
        columnas_variables = columnas[15:22]

        #Columnas fijas 

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'ETAP_PROY', 'JEFE_PROY', 'INST_PROY', 'NOM_SECTOR', 'COD_REG', 'FECHA_INI', 'FECHA_TER', 'N_ESTACION', 'LATITUD', 'LONGUITUD', 'PROF_SECT','PHYLUM', 'CLASE', 'FAMILIA', 'GEN/ESPECIE']

        #Realizamos un melt para convertir las columnas variables a filas 

        df_macrofauna_copy = pd.melt(df_macrofauna_copy , id_vars=columnas_fijas , value_vars=columnas_variables , var_name= 'VARIABLE' , value_name='VALOR')

        #Renombramos las columnas de las Variables 

        #Modificamos el NOMB_PROY

        df_macrofauna_copy['NOMB_PROY'] = 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2020-2021'

        #Modificamos el COD_PROY

        df_macrofauna_copy['COD_PROY'] = '656-130'

        #Modificamos los Phyllum , Clase , Familia , Gen/Especie

        df_macrofauna_copy['PHYLUM'] = np.where(df_macrofauna_copy['PHYLUM'] == '-' , np.nan , df_macrofauna_copy['PHYLUM'])
        df_macrofauna_copy['CLASE'] = np.where(df_macrofauna_copy['CLASE'] == '-' , np.nan , df_macrofauna_copy['CLASE'])
        df_macrofauna_copy['FAMILIA'] = np.where(df_macrofauna_copy['FAMILIA'] == '-' , np.nan , df_macrofauna_copy['FAMILIA'])
        df_macrofauna_copy['GEN/ESPECIE'] = np.where(df_macrofauna_copy['GEN/ESPECIE'] == '-' , np.nan , df_macrofauna_copy['GEN/ESPECIE'])

        #Renombramos las columnas

        df_macrofauna_copy = df_macrofauna_copy.rename(columns={"GEN/ESPECIE" : "TAXA"})

        #Modificamos los formatos de las fechas 
        df_macrofauna_copy['FECHA_INI'] = df_macrofauna_copy['FECHA_INI'].dt.strftime("%d/%m/%Y %H:%M:%S")
        df_macrofauna_copy['FECHA_TER'] = df_macrofauna_copy['FECHA_TER'].dt.strftime("%d/%m/%Y %H:%M:%S")
        
        #Modificamos las columnas de Variable 

        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'N°_TAXA', 'NUMERO DE TAXA' , df_macrofauna_copy['VARIABLE'])
        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'NUM_EJEMPL', 'NUMERO DE EJEMPLARES' , df_macrofauna_copy['VARIABLE'])
        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'TIPO_DRAGA', 'TIPO DRAGA' , df_macrofauna_copy['VARIABLE'])
        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'TIPO_DRAGA', 'TIPO DRAGA' , df_macrofauna_copy['VARIABLE'])
        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'ABUNDANCIA' , 'ABUNDANCIA TOTAL' , df_macrofauna_copy['VARIABLE'])
        df_macrofauna_copy['VARIABLE'] = np.where(df_macrofauna_copy['VARIABLE'] == 'BIOMASA' , 'BIOMASA TOTAL' , df_macrofauna_copy['VARIABLE'])
        
        #Eliminamos las tildes de los SECTORES

        df_macrofauna_copy['NOM_SECTOR'] =  df_macrofauna_copy['NOM_SECTOR'].str.normalize('NFD').str.replace(r'[\u0300-\u036f]', '', regex=True)
                
        return df_macrofauna_copy , columnas , columnas_variables 

    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_macrofauna: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None #O devuelve un valor que indique el fallo
```

```python
df_macrofauna_copy , columnas , columnas_variables  = limpieza_macrofauna(df_macrofauna)
```

# Limpieza de CTD


```python
df_ctd = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2020_P656130\3_BD_CTD_SEGUIMIENTO_2020.xlsx' , sheet_name = 'CTD_2020_IFOP_656-130')
```

```python
df_ctd
```

```python
def limpieza_ctd(df):

    try:

        #Eliminamos las filas innecesarias
        df = df.iloc[18: , :]

        #Creamos una copia del dataset 
        df_ctd_copy = df.copy()

        #Reiniciamos los indices del dataset

        df_ctd_copy = df_ctd_copy.reset_index(drop = True)

        #Dejamos la primera fila como columnas 

        df_ctd_copy.columns = df_ctd_copy.iloc[0]

        #Eliminamos la primera fila del dataset 
        
        df_ctd_copy = df_ctd_copy.iloc[1: , :]

        #Volvemos a reiniciar los indices 

        df_ctd_copy = df_ctd_copy.reset_index(drop = True)

        #Creamos una columna de ID 

        df_ctd_copy.insert(0 , 'ID' , df_ctd_copy.index + 1)

        #Creamos la lista de columnas 

        columnas = df_ctd_copy.columns.to_list()

        #Creamos la lista de columnas fijas 

        columnas_fijas = ['ID', 'NOMB_PROY', 'COD_PROY', 'FECHA_INI', 'FECHA_TER', 'EQUIPO', 'LATITUD', 'LONGUITUD', 'PROF_EQ', 'PROF_SECT' , 'COD_REG', 'OBSERV', 'ESTACION', 'HORA_INICIO' ]

        columnas_variables = ['TEMPERATUR', 'OXIG_ml/L', 'OXIG_mg/L', 'OXIG_%sat', 'OX_umol/kg', 'SALINIDAD', 'FLUORESCEN', 'DENSIDAD', 'CLOROFILA', 'FEOPIGMEN', 'NITRATO', 'FOSFATO', 'SILICATO' , 'DISCO_SECH']

        #Realizamos un melt para convertir las columnas Variables en filas 

        df_ctd_copy = pd.melt(df_ctd_copy , id_vars=columnas_fijas , value_vars=columnas_variables , var_name= 'VARIABLE' , value_name='VALOR')

        # Creamos una funcion para convertir la columna de HORA_INICIO a datetime para luego Crear una columna de fecha y hora 
        def convertir_hora(time_value):
            if isinstance(time_value, str): 
                # Separamos el ":" del texto y lo convertimos a entero y obtenemos dos variables
                horas, minutos = map(int, time_value.split(':'))
                #Retornamos los valores de Horas y minutos
                return datetime.time(horas, minutos)
            else:
                # En caso que ya es datetime.time se deja como esta
                return time_value


        #Aplicamos la funcion de convercion de hora a la columna de HORA_INICIO
         
        df_ctd_copy['HORA_CONVERTIDA'] = df_ctd_copy['HORA_INICIO'].apply(convertir_hora)  

        

        #CREAMOS UNA COLUMNA DE FECHA_INI_CONVERTIDA EN EL FORMATO DE FECHA DE CHILE

        df_ctd_copy['FECHA_INI_CONVERTIDA'] =  pd.to_datetime(df_ctd_copy['FECHA_INI'] , format= '%d-%m-%Y').dt.strftime("%d/%m/%Y")

        #Por ultimo combinamos la HORA_CONVERTIDA Y FECHA_INI_CONVERTIDA 

        df_ctd_copy['FECHA_HORA_COMBINADA_DT'] = df_ctd_copy.apply(
        lambda row: datetime.datetime.strptime(row['FECHA_INI_CONVERTIDA'], "%d/%m/%Y").replace(
            hour=row['HORA_CONVERTIDA'].hour,
            minute=row['HORA_CONVERTIDA'].minute,
            second=row['HORA_CONVERTIDA'].second
        ),
        axis=1
    )

        #Creamos la columna final para la fecha y hora en el formato de chile
        df_ctd_copy['FECHA_HORA_FINAL_STRING'] = df_ctd_copy['FECHA_HORA_COMBINADA_DT'].dt.strftime("%d/%m/%Y %H:%M:%S")
   
        #Eliminamos las columnas innecesarias

        df_ctd_copy = df_ctd_copy[['ID', 'NOMB_PROY', 'COD_PROY', 'FECHA_INI', 'FECHA_TER', 'EQUIPO', 'LATITUD', 'LONGUITUD', 'PROF_EQ', 'PROF_SECT', 'COD_REG', 'OBSERV', 'ESTACION', 'HORA_INICIO', 'VARIABLE', 'VALOR','FECHA_HORA_FINAL_STRING']]


        #Renombramos las columnas de Variable 

        df_ctd_copy['VARIABLE'] = np.where(df_ctd_copy['VARIABLE'] == 'TEMPERATUR' , 'TEMPERATURA' , df_ctd_copy['VARIABLE'])
        df_ctd_copy['VARIABLE'] = np.where(df_ctd_copy['VARIABLE'] == 'FLUORESCEN' , 'FLUORESCENCIA' , df_ctd_copy['VARIABLE'])
        df_ctd_copy['VARIABLE'] = np.where(df_ctd_copy['VARIABLE'] == 'DISCO_SECH' , 'DISCO_SECCHI' , df_ctd_copy['VARIABLE'])

        #Aplicamos un UPPER a la columna de VARIABLE para dejar todo en mayuscula 

        df_ctd_copy['VARIABLE'] = df_ctd_copy['VARIABLE'].str.upper()

        




        return df_ctd_copy , columnas_fijas , columnas_variables

   
    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_ctd: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None #O devuelve un valor que indique el fallo
```

```python
df_ctd_copy , columnas_fijas  , columnas_variables = limpieza_ctd(df_ctd)
```

# Limpieza de Muestreo Regular

```python
df_muestreo_regular = pd.read_excel(r'C:\Users\felipe.abarzua\Desktop\DATA_AMBIENTALES\Proyecto Seguimiento Ambiental\2020_P656130\1_BD_MuestreoRegular_SEGUIMIENTO_2020.xlsx' , sheet_name='BD_MREGULAR2020')
```

```python
df_muestreo_regular
```

```python
def limpieza_muestreo_regular(df):

    try:

        #Eliminamos las filas innecesarias
        df = df.iloc[16: , :]

        #Creamos una copia del dataset 
        df_muestreo_regular_copy = df.copy()

        #Reiniciamos los indices del dataframe 
        df_muestreo_regular_copy = df_muestreo_regular_copy.reset_index(drop = True)

        #Dejamos la primera fila como columnas 
        df_muestreo_regular_copy.columns = df_muestreo_regular_copy.iloc[0]

        #Eliminamos la primera fila y luego reiniciamos los indices 
        df_muestreo_regular_copy = df_muestreo_regular_copy.iloc[1: , :]

        #Reiniciamos los indices del dataframe 
        df_muestreo_regular_copy = df_muestreo_regular_copy.reset_index(drop = True)

        #Reemplazamos los valores de -- en HORA_INI y HORA_FIN por datetime.time(0,0)

        df_muestreo_regular_copy['HORA_INI'] = np.where(df_muestreo_regular_copy['HORA_INI'] == '--' , datetime.time(0,0) , df_muestreo_regular_copy['HORA_INI'])
        df_muestreo_regular_copy['HORA_FIN'] = np.where(df_muestreo_regular_copy['HORA_FIN'] == '--' , datetime.time(0,0) , df_muestreo_regular_copy['HORA_FIN'])
        df_muestreo_regular_copy['HORA_FIN'] = np.where(df_muestreo_regular_copy['HORA_FIN'] == '-' , datetime.time(0,0) , df_muestreo_regular_copy['HORA_FIN'])
        
        #Creamos una columna ID 
        df_muestreo_regular_copy.insert(0 , "ID" , df_muestreo_regular_copy.index + 1)

        #Obtenemos la lista de columnas 
        columnas = df_muestreo_regular_copy.columns.to_list()

        #Creamos la lista de columnas fijas y de variables
        columnas_fijas = columnas[:17]
        columnas_variables = columnas[17:38]
        
        #Realizamos un melt para convertir las columnas de variables a filas 

        df_muestreo_regular_copy = pd.melt(df_muestreo_regular_copy , id_vars=columnas_fijas , value_vars=columnas_variables , var_name='VARIABLE_MUESTREO' , value_name='VALOR')

        #Combinamos las fechas utlizando la fecha y Hora 

        # Combine FECHA and Hora columns
        df_muestreo_regular_copy['FECHA_INICIO'] = df_muestreo_regular_copy.apply(lambda row: datetime.datetime.combine(row['FECHA_INI'].date(), row['HORA_INI']), axis=1)
        df_muestreo_regular_copy['FECHA_TERMINO'] = df_muestreo_regular_copy.apply(lambda row : datetime.datetime.combine(row['FECHA_TER'].date(), row['HORA_FIN']), axis=1)

        #Reemplazamos los valores de sd por na
        df_muestreo_regular_copy['VALOR'] = np.where(df_muestreo_regular_copy['VALOR'] == 'sd' , np.nan , df_muestreo_regular_copy['VALOR'])
        df_muestreo_regular_copy['VALOR'] = np.where(df_muestreo_regular_copy['VALOR'] == '--' , np.nan , df_muestreo_regular_copy['VALOR'])

        #Cambiamos los nombres de Proyectos
        df_muestreo_regular_copy['NOMB_PROY'] = np.where(df_muestreo_regular_copy['NOMB_PROY'] == "ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2020 -2021" , "ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2020-2021" , df_muestreo_regular_copy['NOMB_PROY'])


        #Dejamos la columna VARIABLE EN MAYUSCULA 

        df_muestreo_regular_copy['VARIABLE_MUESTREO'] = df_muestreo_regular_copy['VARIABLE_MUESTREO'].str.upper()

        
        # Modificamos los formatos de las FECHAS
        df_muestreo_regular_copy['FECHA_INICIO'] = df_muestreo_regular_copy['FECHA_INICIO'].dt.strftime('%d/%m/%Y %H:%M:%S')
        df_muestreo_regular_copy['FECHA_TERMINO'] = df_muestreo_regular_copy['FECHA_TERMINO'].dt.strftime('%d/%m/%Y %H:%M:%S')

      
        #Cambiamos los nombres de la VARIABLE_MUESTREO

        df_muestreo_regular_copy['VARIABLE_MUESTREO'] = np.where(df_muestreo_regular_copy['VARIABLE_MUESTREO'] == 'BIOM_TOTAL' , 'BIOMASA_TOTAL' , df_muestreo_regular_copy['VARIABLE_MUESTREO'] )




        return df_muestreo_regular_copy , columnas , columnas_fijas , columnas_variables


    except Exception as e:
        print(f"Ocurrió un error en la función limpieza_muestreo_regular: {e}")
        ex = sys.exc_info()[2]
        print("Error en el archivo : " + str(ex.tb_frame) + " En la linea : " + str(ex.tb_lineno))
        
    return None

```

```python
df_muestreo_regular_copy , columnas , columnas_fijas , columnas_variables = limpieza_muestreo_regular(df_muestreo_regular)
```
