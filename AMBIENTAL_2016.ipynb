# Creamos la funcion para poder limpiar la data y dejarla en el formato deseado

```python
def limpieza_macrofauna(df):

    #Eliminamos las columnas innecesarias 

    df = df.iloc[15: ,  :-3]

    #Realizamos una copia del dataset 
    df_macrofauna_copy = df.copy()

    #Reiniciamos los indices del dataframe 

    df_macrofauna_copy =  df_macrofauna_copy.reset_index()

    #Eliminamos la columna de index 

    df_macrofauna_copy = df_macrofauna_copy.iloc[: , 1:]

    #Luego dejamos la primera fila como columna 

    df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

    #Dejamos la primera fila como columnas 
    df_macrofauna_copy.columns = df_macrofauna_copy.iloc[0]

     #Eliminamos la primera fila 
    df_macrofauna_copy =  df_macrofauna_copy.iloc[1: , :]

    # Creamos una lista de las columnas
    columnas = df_macrofauna_copy.columns.tolist()

    # Cambiamos los nombres de las columnas desde la posicion 17 hasta la 20

    columnas[17] = 'AMBI INV'
    columnas[18] = 'CLASIFICACIÓN DE PERTURBACIÓN INV'
    columnas[19] = 'AMBI VER'
    columnas[20] =  'CLASIFICACIÓN DE PERTURBACIÓN VER'

    # Realizamos dos ciclos para modificar las columnas de variables para agregarle INV o VER

    for indice , columna in enumerate(columnas):

        if indice >= 7 and indice < 17:

            print(f"La columna '{columna}' esta en la posicion {indice}")

            if indice % 2 == 0:

                print(f"La columna '{columna} se debe agregar '{columnas[indice - 1]}' y VER para convertirse en  '{columnas[indice - 1]} VER' ")

                columnas[indice] = columnas[indice - 1] + ' VER' 


    for indice , columna in enumerate(columnas):

        if indice >= 7 and indice < 17:

            print(f"La columna '{columna}' esta en la posicion {indice}")

            if indice % 2 != 0:

                print(f"La columna '{columna} se debe agregar INV para convertirse en  '{columnas[indice]} INV ' ")

                columnas[indice] = columnas[indice] + ' INV' 


    
    # Utilizamos expresiones regulares para eliminar los espacios en blanco adicionales
    columnas = [re.sub(r'\s+' , ' ' , columna) for columna in columnas]
    
    #Modificamos los nombres de las columnas del dataframe 
    df_macrofauna_copy.columns = columnas 

    #Eliminamos la primera fila del dataframe 
    df_macrofauna_copy = df_macrofauna_copy.iloc[1: , :]

    #Reiniciamos los indices del dataframe 
    df_macrofauna_copy = df_macrofauna_copy.reset_index()

    #Eliminamos la columna de index creada 
    df_macrofauna_copy = df_macrofauna_copy.iloc[: ,  1:]

    #Agregamos una columna de id 
    df_macrofauna_copy.insert(0 ,  'ID' , df_macrofauna_copy.index + 1)
    
    # Volvemos a crear las columnas
    columnas = df_macrofauna_copy.columns.tolist()

    #Creamos las columnas fijas y las columnas de variables para luego realizar el melt 
    columnas_fijas = columnas[:8]
    columnas_variables = columnas[8:]


    # Realizamos un melt para convertir las columnas de variables 
    df_macrofauna_copy = pd.melt(df_macrofauna_copy ,  id_vars = columnas_fijas , var_name= 'VARIABLE' , value_vars= columnas_variables , value_name = 'VALOR')

    
    # Creamos una funcion para extraer los INV o VER de las Variables 

    def obtener_ultimo_tres(texto):
        if isinstance(texto, str) and len(texto) >= 3:
            return texto[-3:]
        else:
            return None  # O algún otro valor por defecto si la cadena es muy corta o no es string

    # Aplica la función a la columna y crea una nueva columna
    df_macrofauna_copy['ESTACION_DEL_ANIO'] = df_macrofauna_copy['VARIABLE'].apply(obtener_ultimo_tres)

     # Creamos una funcion para eliminar los INV y VER de la columna VAriable 
    def eliminar_inv_ver(texto):

        if isinstance(texto, str):
         texto = texto.replace('INV', '')
         texto = texto.replace('VER', '')
        return texto


    #Aplicamos la funcion para eliminar los INV y VER de la columna VARIABLE
    df_macrofauna_copy['VARIABLE'] = df_macrofauna_copy['VARIABLE'].apply(eliminar_inv_ver)

    #Eliminamos los espacios en blanco al final del texto 
    df_macrofauna_copy['ZONA'] = df_macrofauna_copy['ZONA'].str.rstrip()
    df_macrofauna_copy['VARIABLE'] = df_macrofauna_copy['VARIABLE'].str.rstrip()

    # Volvemos a crear las columnas
    columnas = df_macrofauna_copy.columns.tolist()

    # Creamos una columna de Unidad 
    df_macrofauna_copy.insert(9 , 'UNIDAD' , np.nan)

     #Modificamos la columna UNIDAD 
    df_macrofauna_copy['UNIDAD'] =  np.where(df_macrofauna_copy['VARIABLE'] ==  'ABUNDANCIA TOTAL (ind/m2)' , 'ind/m2' ,  df_macrofauna_copy['UNIDAD']  )
    df_macrofauna_copy['UNIDAD'] =  np.where(df_macrofauna_copy['VARIABLE'] ==  'BIOMASA TOTAL(g/m2)' , 'g/m2' ,  df_macrofauna_copy['UNIDAD']  )

    # Modificamos la columna VARIABLE 
    df_macrofauna_copy['VARIABLE'] =  np.where(df_macrofauna_copy['VARIABLE'] ==  'ABUNDANCIA TOTAL (ind/m2)' , 'ABUNDANCIA TOTAL' ,  df_macrofauna_copy['VARIABLE'] )
    df_macrofauna_copy['VARIABLE'] =  np.where(df_macrofauna_copy['VARIABLE'] ==  'BIOMASA TOTAL(g/m2)' , 'BIOMASA TOTAL' ,  df_macrofauna_copy['VARIABLE'] )

    # Remplazamos los valores sd de la columna valor 

    df_macrofauna_copy['VALOR'] = np.where(df_macrofauna_copy['VALOR'] == 'sd' , np.nan , df_macrofauna_copy['VALOR'])

     #Creamos un diccionario para poder mapear la ZONA con el COD_REGION 

    mapeo_regiones = {'Estuario' : 'X' , 'Reloncavi' : 'X' , 'Calbuco' : 'X' , 'Maullin' : 'X' , 'Ancud' : 'X' , 'Chacao' : 'X' , 'Butachauques' : 'X' , 'Chaulinec' : 'X' , 'Tenaun' : 'X' , 'Chiloe central'  : 'X' ,
                  'Desertores' : 'X' , 'Quellon' : 'X' ,  'Hornopiren' : 'X' , 'Ayacara' : 'X' , 'Chaiten' : 'X' ,  'Guaitecas' : 'XI' , 'Moraleda' : 'XI' , 'Puyuhuapi' : 'XI' , 'Aysen' : 'XI' , 'Coquimbo' : 'IV' , 'Atacama' : 'III' , 'Pto. Natales' : 'XII' , 'Bio-Bio' : 'VIII' , 
                  'Valdivia' : 'XIV'}

    # Creamos la columna de COD_REGION
    df_macrofauna_copy['COD_REGION'] = df_macrofauna_copy['ZONA'].map(mapeo_regiones)

    #Agregamos la columna de Año

    df_macrofauna_copy.insert(1 , 'AÑO' , 2016)

    #Modificamos el NOMB_PROY

    df_macrofauna_copy['NOMB_PROY'] = np.where(df_macrofauna_copy['NOMB_PROY'] == 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2016 -2017' , 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2016-2017' , df_macrofauna_copy['NOMB_PROY'])


    return df_macrofauna_copy ,  columnas , columnas_fijas , columnas_variables



```

```python
def limpieza_ctd(df):

    #Eliminamos las filas innecesarias 

    df =  df.iloc[10: , :]

    #Realizamos copia del dataframe
    df_ctd_copy = df.copy()

    #Reiniciamos los index 
    df_ctd_copy =  df_ctd_copy.reset_index()

    #Eliminamos la columna de index 
    df_ctd_copy = df_ctd_copy.iloc[: , 1:]

    #Dejamos la primera fila como columnas 
    df_ctd_copy.columns = df_ctd_copy.iloc[0]

    #Eliminamos la primera fila 
    df_ctd_copy = df_ctd_copy.iloc[2: , :]  

    #Reiniciamos los index 
    df_ctd_copy =  df_ctd_copy.reset_index()

    #Eliminamos la columna de index 
    df_ctd_copy = df_ctd_copy.iloc[: , 1:]

    #Creamos una columna de ID 
    df_ctd_copy.insert(0 , 'ID' , df_ctd_copy.index + 1  )
    
    #Renombramos las columnas
    df_ctd_copy = df_ctd_copy.rename(columns= { 'STATION' : 'ESTACION'  , 'CRUISE'  : 'SEGUIMIENTO' , 'LATITUDE' : 'LATITUD' , 'LONGITUDE' : 'LONGITUD'  ,  'TEMPERATURE_(°C)' : 'TEMPERATURA °C' ,  'OXIGENO_DIS_(ML/L)' : 'OXIGENO_DIS_ML/L ML/L' , 'OXIGENO_DIS_(MG/L)' : 'OXIGENO_DIS_MG/L MG/L' , 'OXIGENO_DIS_(%SAT)' : 'OXIGENO_DIS_SAT %' , 'OXIGENO_DIS_(UMOL/KG)' : 'OXIGENO_DIS_UMOLKG UMOL/KG' , 'SALINITY_(PSU)' : 'SALINIDAD PSU' , 'DENSIDAD_(KG/M3)' : 'DENSIDAD KG/M3' , 'FLUORESCENCIA_(MG/M-3)' : 'FLUORESCENCIA MG/M-3' , 'DEPHT_(M)' : 'PROFUNDIDAD'})

    # Creamos una columna de FECHA en la posicion 9 del dataframe
    df_ctd_copy.insert(10 , 'FECHA' , df_ctd_copy.apply(lambda row: datetime.datetime(row['YEAR'], row['MONTH'], row['DAY']), axis=1) )  

    #Formateamos la FECHA  a DD/MM/YYYY 
    # Formatear la fecha y hora al formato deseado
    df_ctd_copy['FECHA'] = df_ctd_copy['FECHA'].dt.strftime('%d/%m/%Y %H:%M:%S')

    #Creamos una lista con las columnas del dataframe
    columnas = df_ctd_copy.columns.tolist()

    #Obtenemos las columnas fijas y las columnas Variables 
    columnas_fijas = columnas[:12]
    columnas_variables = columnas[12:]

    #Realizamos un melt para convertir las columnas variables en filas
    df_ctd_copy = pd.melt(df_ctd_copy , id_vars= columnas_fijas , var_name= 'VARIABLE' ,  value_vars = columnas_variables , value_name= 'VALOR')

    #Creamos una lista con las columnas del dataframe
    columnas = df_ctd_copy.columns.tolist()

    #Creamos una columna llamada Unidad obteniendo lo que viene despues del espacio en la columna VARIABLE
    df_ctd_copy.insert(13 , 'UNIDAD' , df_ctd_copy['VARIABLE'].str.split(' ', n=1).str[1])

    #Luego eliminamos las Unidades de la columna Variable 
    df_ctd_copy['VARIABLE'] =  df_ctd_copy['VARIABLE'].str.split(' ' , n=1).str[0]

    #Dejamos todas las estaciones en MAYUSCULA
    df_ctd_copy['ESTACION'] =  df_ctd_copy['ESTACION'].str.upper()

    #Reemplazamos los NOMB_PROY 
    df_ctd_copy['NOMB_PROY'] =  np.where(df_ctd_copy['NOMB_PROY'] == 'ESTUDIO DEL DESEMPEÑO AMBIENTAL DE LA ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO, 2016-2017' , 'ESTUDIO DEL DESEMPEÑO AMBIENTAL DE LA ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2016-2017' , df_ctd_copy['NOMB_PROY'])

    #Corregimos los valores de las estaciones 
      # Crear un diccionario de mapeo para tus asignaciones
    # Creamos un diccionario de mapeo de sectores
    
    mapeo_sectores = {
        
            "TEN" :  "Tenaun" ,
            "CHAU" : "Chaulinec" ,
            "CHIC" : "Chiloe central" ,
            "CHAC" : "Chacao" ,
            "QUELL" : "Quellon",
            "BUT" : "Butachauques",
            "ANC" : "Ancud" ,
            "REL" : "Reloncavi" ,
            "CALB" : "Calbuco" ,
            "HORN" : "Hornopiren" , 
            "VALD" : "Valdivia" ,
            "AYA" : "Ayacara" 
    }


    mapeo_regiones = {'Estuario' : 'X' , 'Reloncavi' : 'X' , 'Calbuco' : 'X' , 'Maullin' : 'X' , 'Ancud' : 'X' , 'Chacao' : 'X' , 'Butachauques' : 'X' , 'Chaulinec' : 'X' , 'Tenaun' : 'X' , 'Chiloe central'  : 'X' ,
                  'Desertores' : 'X' , 'Quellon' : 'X' ,  'Hornopiren' : 'X' , 'Ayacara' : 'X' , 'Chaiten' : 'X' ,  'Guaitecas' : 'XI' , 'Moraleda' : 'XI' , 'Puyuhuapi' : 'XI' , 'Aysen' : 'XI' , 'Coquimbo' : 'IV' , 'Atacama' : 'III' , 'Pto. Natales' : 'XII' , 'Bio-Bio' : 'VIII' , 
                  'Valdivia' : 'XIV'}

    # Extraemos los Números de la columna ESTACION para crear una la columna N_ESTACION
    df_ctd_copy['N_ESTACION'] = df_ctd_copy['ESTACION'].str.extract(r'(\d+)')

    #Obtenemos el código utilizando la expresion regular 
    df_ctd_copy['CODIGO_EXTRAIDO'] = df_ctd_copy['ESTACION'].str.extract(r'([A-Z]{3,5})')

    #Luego utilizando el diccionario de mapeo creamos la columna de SECTOR utilizando la funcion map

    df_ctd_copy['SECTOR'] = df_ctd_copy['CODIGO_EXTRAIDO'].map(mapeo_sectores)

    #Utilizamos el diccionario de mapeo de regiones para crear la columna COD_REGION

    df_ctd_copy['COD_REGION'] = df_ctd_copy['SECTOR'].map(mapeo_regiones)

    #Se van a eliminar los registros que están nulos en COD_REGION

    df_ctd_copy = df_ctd_copy.dropna(subset=['COD_REGION'])

    #Cambiamos el tipo de dato de la columna N_ESTACION
    df_ctd_copy['N_ESTACION'] = df_ctd_copy['N_ESTACION'].astype(int)

    return df_ctd_copy

```

```python

def limpieza_fisicoquimico(df):

    df_fisio_copy = df_fisio.copy()
    
    #Eliminamos las filas innecesarias
    df_fisio_copy =  df_fisio_copy.iloc[14: , :]

    # Reiniciamos los index
    df_fisio_copy = df_fisio_copy.reset_index(drop= True)

    #Dejamos la primera fila como columna 
    df_fisio_copy.columns = df_fisio_copy.iloc[0]

    # Creamos una lista de las columnas
    columnas = df_fisio_copy.columns.tolist()

    #Eliminamos la primera fila del dataset
    df_fisio_copy = df_fisio_copy.iloc[1: , :]

    # Reiniciamos los index
    df_fisio_copy =  df_fisio_copy.reset_index(drop= True)

    # Eliminamos las columnas innecesarias

    df_fisio_copy = df_fisio_copy.iloc[:, 
        list(range(0, 4)) +        # 0 a 3
        list(range(5, 12)) +       # 5 a 11
        [13, 15, 17, 19, 21, 23, 25, 27] +  # índices sueltos
        list(range(28, 39))        # 28 a 39
    ]

    # Realizamos una lista de las columnas 

    lista_columnas = df_fisio_copy.columns.tolist()




    # Realizamos un ciclo para poder modificar las columnas del dataframe

    for indice, columna in enumerate(lista_columnas):
        if 10 <= indice < 30:
            if indice % 2 == 0:
                # Par: agregar INV al nombre actual
                lista_columnas[indice] = f"{columna} INV"
            else:
                # Impar: tomar la anterior y quitar ' INV' si lo tiene
                prev = lista_columnas[indice - 1]
                if isinstance(prev, str) and prev.endswith(' INV'):
                    prev_base = prev[:-4]
                else:
                    prev_base = prev
                lista_columnas[indice] = f"{prev_base} VER"



    # Creamos una funcion para eliminar todo el contenido que son (*)

    def es_nan(x):
        return isinstance(x, float) and math.isnan(x)


    def quitar_parentesis_contenido(lista):
    
        resultado = []
        patron = re.compile(r'\([^)]*\)')  # todo lo que esté entre paréntesis (no anida)
        for c in lista:
            if es_nan(c) or c is None:
                resultado.append(c)
            elif isinstance(c, str):
                s = patron.sub('', c)                 # quita ( ... )
                s = re.sub(r'\s{2,}', ' ', s).strip() # colapsa espacios y recorta
                resultado.append(s)
            else:
                resultado.append(c)
        return resultado


    # Utilizamos la funcion para limpiar la lista de columnas 

    lista_columnas = quitar_parentesis_contenido(lista_columnas)


    # Asignamos la lista de columnas creadas al dataframe 

    df_fisio_copy.columns = lista_columnas


    # Eliminamos la primera fila 

    df_fisio_copy = df_fisio_copy.iloc[1: , :]

    # Reiniciamos los index
    df_fisio_copy =  df_fisio_copy.reset_index(drop= True)

    # Crear la columna ID autoincrementable
    id_columna = range(1, len(df_fisio_copy) + 1)  # genera 1, 2, 3, ...

    # Agregamos una columna de ID 
    df_fisio_copy.insert(0 , "ID" , df_fisio_copy.index + 1 )

    # Creamos dos dataset uno para INV y VER

    df_fisio_copy_inv = df_fisio_copy[['ID', 'NOMB_PROY' ,'ETAP_PROY', 'JEFE_PROY', 'INST_PROY', 'ZONA', 'ESTACION',
        'Latitud S INV' , 'Longitud W INV' , 'pH INV', 'Temperatura INV',
        'Redox EHNHE INV',  'Sulfuros INV', 
        'Materia orgánica total INV', 'Arena INV',
        'Fango INV',  'Grava INV',
        'Tamaño medio INV', 'Selección INV']]


    # Creamos la lista de columnas para el dataset inv

    lista_columnas_inv = df_fisio_copy_inv.columns.to_list()


    df_fisio_copy_ver = df_fisio_copy[['ID', 'NOMB_PROY' , 'ETAP_PROY', 'JEFE_PROY', 'INST_PROY', 'ZONA', 'ESTACION',
        'Latitud S VER', 'Longitud W VER', 'pH VER',  'Temperatura VER',
        'Redox EHNHE VER', 'Sulfuros VER',
        'Materia orgánica total VER', 
        'Arena VER', 'Fango VER',  'Grava VER',
            'Tamaño medio VER', 'Selección VER']]

    # Creamos la lista de columnas para el dataset ver 

    lista_columnas_ver = df_fisio_copy_ver.columns.to_list()

    #  Realizar un melt a cada dataset para convertir las columnas en filas 

    def convertir_columnas_dataframe(df , lista_columnas):
    
    # Obtenemos las columnas fijas y variables
        columnas_fijas = [lista_columnas[i] for i in range(0, 9)]      # posiciones 0 a 8
        columnas_variables = [lista_columnas[i] for i in range(9, 18)] # posiciones 8 a 17

    # Realizamos un melt para convertir las columnas variables en filas

        df = df.melt( id_vars = columnas_fijas , var_name = "VARIABLE"  , value_vars = columnas_variables , value_name = 'VALOR')

        return df , columnas_fijas , columnas_variables

    # Realizamos el melt de INV y VER para convertir las columnas de variables en filas
    df_fisio_inv_melt , columnas_fijas_inv , columnas_variables_inv = convertir_columnas_dataframe(df_fisio_copy_inv , lista_columnas_inv)
    df_fisio_ver_melt , columnas_fijas_ver , columnas_variables_ver = convertir_columnas_dataframe(df_fisio_copy_ver , lista_columnas_ver)

    # Reemplazamos los valores de sd de la columna VALOR por nulo 

    df_fisio_ver_melt['VALOR'] = np.where(df_fisio_ver_melt["VALOR"] == "sd" , np.nan , df_fisio_ver_melt['VALOR'] )
    df_fisio_inv_melt['VALOR'] = np.where(df_fisio_inv_melt["VALOR"] == "sd" , np.nan , df_fisio_inv_melt['VALOR'] )

    # Creamos una columna llamada ESTACION_DEL_ANIO para diferenciar INV Y VER

    df_fisio_ver_melt['ESTACION_DEL_ANIO'] = "VER" 
    df_fisio_inv_melt['ESTACION_DEL_ANIO'] = "INV"


    # Creamos una funcion para eliminar los grados , minutos y segundos con regex

    # Aplicamos la funcion a la latitud y longitud  a los dataframes

    df_fisio_ver_melt['Longitud W VER'] = df_fisio_ver_melt['Longitud W VER'].str.replace(r"[°'\"]", "", regex=True)
    df_fisio_ver_melt['Latitud S VER'] = df_fisio_ver_melt['Latitud S VER'].str.replace(r"[°'\"]", "", regex=True)

    df_fisio_inv_melt['Longitud W INV'] = df_fisio_inv_melt['Longitud W INV'].str.replace(r"[°'\"]", "", regex=True)
    df_fisio_inv_melt['Latitud S INV'] = df_fisio_inv_melt['Latitud S INV'].str.replace(r"[°'\"]", "", regex=True)

    # Borrar INV y VER de los campos LATITUD Y LONGITUD y VARIABLE

    df_fisio_inv_melt = df_fisio_inv_melt.rename(columns={'Latitud S INV' : 'LATITUD' , 'Longitud W INV' : 'LONGITUD'})

    df_fisio_ver_melt =  df_fisio_ver_melt.rename(columns={'Latitud S VER' : 'LATITUD' , 'Longitud W VER' : 'LONGITUD' })

    # Ahora borramos INV y VER de las VARIABLES

    df_fisio_inv_melt['VARIABLE'] = df_fisio_inv_melt['VARIABLE'].str.replace('INV' , '')

    df_fisio_ver_melt['VARIABLE'] = df_fisio_ver_melt['VARIABLE'].str.replace('VER' , '')

    # Luego aplicamos un rstrip para eliminar los espacios en blanco al final del texto

    df_fisio_inv_melt['VARIABLE'] = df_fisio_inv_melt['VARIABLE'].str.rstrip()
    df_fisio_ver_melt['VARIABLE'] = df_fisio_ver_melt['VARIABLE'].str.rstrip()

    # Ahora aplicamos un upper para dejar todo en mayúscula
    df_fisio_inv_melt['VARIABLE'] = df_fisio_inv_melt['VARIABLE'].str.upper()
    df_fisio_ver_melt['VARIABLE'] = df_fisio_ver_melt['VARIABLE'].str.upper()

    # Ahora habia que realizar un concat para unir los dataframes 

    df_fisio_2016 = pd.concat([df_fisio_inv_melt , df_fisio_ver_melt])

    # Reemplazamos los valores de las variables

    df_fisio_2016['VARIABLE'] = np.where(df_fisio_2016['VARIABLE'] == 'MATERIA ORGÁNICA TOTAL' , 'MATERIA ORGANICA' , df_fisio_2016['VARIABLE'])


    #Eliminamos los espacios en blanco de la latitud y longitud 
    df_fisio_2016['LATITUD'] = df_fisio_2016['LATITUD'].str.replace(' ' , '')
        
    df_fisio_2016['LONGITUD'] = df_fisio_2016['LONGITUD'].str.replace(' ' , '')


    #Realizamos un replace para convertir los , por . 

    df_fisio_2016['LATITUD'] = df_fisio_2016['LATITUD'].str.replace(',' , '.')
    df_fisio_2016['LONGITUD'] = df_fisio_2016['LONGITUD'].str.replace(',' , '.')

    #Convertimos la LATITUD y LONGITUD en float

    df_fisio_2016['LATITUD'] = df_fisio_2016['LATITUD'].astype(float)
    df_fisio_2016['LONGITUD'] = df_fisio_2016['LONGITUD'].astype(float)


    # Creamos una funcion para crear los COD_REGION

    
    def normalizar(s):
        if pd.isna(s):
            return s
        # quitar tildes, bajar a minúsculas y recortar espacios
        s = ''.join(c for c in unicodedata.normalize('NFD', str(s).strip().lower())
                    if unicodedata.category(c) != 'Mn')
        return s

# Agregamos la columna de COD_REGION
   

# Diccionario en números romanos
    
    
    mapa_romano = {
        # Los Lagos (X)
        'reloncavi': 'X',
        'estuario': 'X',
        'calbuco': 'X',
        'maullin': 'X',
        'ancud': 'X',
        'chacao': 'X',
        'butachauques': 'X',
        'chaulinec': 'X',
        'tenaun': 'X',
        'chiloe central': 'X',
        'desertores': 'X',
        'quellon': 'X',
        'hornopiren': 'X',
        'ayacara': 'X',
        'chaiten': 'X',

        # Aysén (XI)
        'guaitecas': 'XI',
        'moraleda': 'XI',
        'puyuhuapi': 'XI',
        'aysen': 'XI',

        # Coquimbo (IV)
        'coquimbo': 'IV',
        'tongoy': 'IV',

        # Atacama (III)
        'atacama': 'III',

        # Magallanes (Natales) – XII
        'natales': 'XII',

        # Biobío (VIII)
        'bio-bio': 'VIII',

        # Los Ríos (XIV)
        'valdivia': 'XIV',
    }




    # Crear columna
   
    df_fisio_2016['COD_REGION'] = df_fisio_2016['ZONA'].apply(normalizar).map(mapa_romano)
        # (Opcional) marca zonas no reconocidas
    df_fisio_2016['COD_REGION'] = df_fisio_2016['COD_REGION'].fillna('DESCONOCIDO')

        #Convertimos la columna de NOMB_PROY
    df_fisio_2016['NOMB_PROY'] = 'ESTUDIO SEGUIMIENTO DESEMPEÑO AMBIENTAL ACUICULTURA EN CHILE Y SU EFECTO EN LOS ECOSISTEMAS DE EMPLAZAMIENTO 2016-2017'

    #Agregamos el año 

    df_fisio_2016['ANIO'] = 2016


    return df_fisio_2016
```
